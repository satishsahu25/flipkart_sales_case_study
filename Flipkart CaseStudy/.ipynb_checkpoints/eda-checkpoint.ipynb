```python
import numpy as np
import pandas as pd
```

```python
import plotly.express as px
```

```python
df=pd.read_csv('flipkart.csv')
df
```

```python
df.tail()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# original price mein jyda null hai so lets drop it
newdf=df.drop(columns=['Original Price'])
```

```python
newdf.isnull().sum()
```

```python
newdf
```

```python
# lets drop the rows having null values
dfclean=newdf.dropna(axis=0)
dfclean
```

```python
dfclean.to_csv('cleanedata.csv',index=False)
```

```python
dfclean.head()
```

```python
# grouping brand wise
brands=dfclean.groupby(by='Brand').count().sort_values(ascending=False,by='Model')
```

```python
brands.head()
```

```python
# lets reset the index
brands=brands.reset_index()
brands.head()
```

# PLOTTING

```python
fig=px.bar(brands,x='Model',y='Brand',color='Brand',title="Number of Models by brands")
```

```python
fig.show()
```

```python
df=dfclean
```

```python
df['Color']=df['Color'].str.split(n=1).str[-1]
```

```python
Colors=df.groupby(by='Color').count().sort_values(ascending=False,by='Model')
Colors=Colors.reset_index()
```

```python
Colors.head()
```

```python
color_10=Colors.sort_values(ascending=False,by=['Model'])[:10]  #keval 10 hi plot karna hai
color_10
```

```python
fig=px.bar(color_10,x='Color',color='Brand',y='Model',title='Top 10 in-Demand Colors')
fig.show()
```

```python
ratings=df.groupby('Brand')[['Rating']].mean().sort_values(ascending=False,by='Rating').reset_index()
```

```python
ratings.head()
```

```python
fig=px.bar(ratings,x='Rating',y='Brand',title='Average rating per brands',color='Brand')
fig.show()
```

```python
fig=px.box(df,x='Brand',y='Selling Price',title='Selling price per brand',color='Brand')
fig.show()
```

```python
# Lets divide by segment of price
df['Segments']=pd.cut(df['Selling Price'],[0,25000,50000,np.inf],labels=['Low Budget','Mid-range','Expensive'])
df.head()
```

```python
segments=df.groupby(['Segments']).count().reset_index()
segments.head()
```

```python
fig=px.pie(segments,values='Model',names='Segments',title='Budget Division')
fig.show()
```

```python
fig=px.scatter(df,x='Rating',y='Selling Price',color='Brand')
fig.show()
```

```python
fig=px.scatter(df,x='Rating',y='Segments',color='Brand')
fig.show()
```
